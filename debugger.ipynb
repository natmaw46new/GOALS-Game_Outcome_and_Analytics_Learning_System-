```python
from pathlib import Path
import pandas as pd

print('=== player_stats.parquet match counts per season ===\n')

for season_dir in sorted(Path('data/47').iterdir()):
    stats_path    = season_dir / 'output' / 'player_stats.parquet'
    fixtures_path = season_dir / 'output' / 'fixtures.parquet'

    if stats_path.exists():
        stats    = pd.read_parquet(stats_path)
        scraped  = stats['match_id'].nunique()
        print(f'{season_dir.name}:')
        print(f'  Unique matches in player_stats : {scraped}')

        if fixtures_path.exists():
            fixtures = pd.read_parquet(fixtures_path)
            finished = fixtures[fixtures['finished'] == True]
            print(f'  Finished fixtures              : {len(finished)}')
            missing  = set(finished['match_id']) - set(stats['match_id'].unique())
            print(f'  Missing matches                : {len(missing)}')
            if missing:
                print(f'  Sample missing IDs             : {list(missing)[:5]}')
        print()
    else:
        print(f'{season_dir.name}: player_stats.parquet NOT FOUND\n')
```
